```python


# Assuming you have a world map GeoDataFrame loaded as 'world'
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
merged_df = world.merge(df, left_on='name', right_on='Entity', how='left')

fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged_df.plot(column='Terrorism deaths per 100,000 people', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
plt.title('Geographical Distribution of Terrorism Deaths')
plt.show()
```
